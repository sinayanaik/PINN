### Define Fully Connected Neural Network (FCN) using PyTorch

- N_INPUT: Number of input features.
- N_OUTPUT: Number of output neurons.
- N_HIDDEN: Number of neurons in each hidden layer.
- N_LAYERS: Number of hidden layers.

- `fcs` : Input Layer (`N_INPUT to N_HIDDEN`) followed by a Tanh activation
- `fch` : Hidden layer Sequence of hidden layers (`N_LAYERS-1`), each with N_HIDDEN neurons and Tanh activation
- `fce` : Output Layer Output layer (`N_HIDDEN to N_OUTPUT`), no activation (linear output)

- Forward Pass (`forward`)



```python
class FCN(nn.Module):
    "Defines a connected network"
    
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS):
        super().__init__()
        activation = nn.Tanh
        self.fcs = nn.Sequential(*[
                        nn.Linear(N_INPUT, N_HIDDEN),
                        activation()])
        self.fch = nn.Sequential(*[
                        nn.Sequential(*[
                            nn.Linear(N_HIDDEN, N_HIDDEN),
                            activation()]) for _ in range(N_LAYERS-1)])
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)
        
    def forward(self, x):
        x = self.fcs(x)
        x = self.fch(x)
        x = self.fce(x)
        return x
```

### Generate Training DATA from anaytical solution

The general equation for a damped harmonic oscillator is given by:

$$
m \ddot{x} + c \dot{x} + kx = 0
$$

where:

- $m$ is the mass,  
- $c$ is the damping coefficient,  
- $k$ is the spring constant.  

For an underdamped system ($ d < w_0 $), the solution is:

$$
y(x) = A e^{-dx} \cos(wx + \phi)
$$

where:

- $d = \frac{c}{2m}$ = damping coefficient,

- $w_0 = \sqrt{\frac{k}{m}}$ = the natural frequency,

- $w = \sqrt{w_0^2 - d^2}$ = damped frequency,

- $\phi$ = phase shift.


```python
def oscillator(d, w0, x):
    """Defines the analytical solution to the 1D underdamped harmonic oscillator problem. 
    """
    assert d < w0
    w = np.sqrt(w0**2-d**2)
    phi = np.arctan(-d/w)
    A = 1/(2*np.cos(phi))
    cos = torch.cos(phi+w*x)
    sin = torch.sin(phi+w*x)
    exp = torch.exp(-d*x)
    y  = exp*2*A*cos
    return y
```

```python
d, w0 = 2, 20

# get the analytical solution over the full domain
x = torch.linspace(0,1,500).view(-1,1) #linspace(start,stop,no of points) , reshapes the (500,) tensor into a (500, 1) tensor.
y = oscillator(d, w0, x).view(-1,1)
print(x.shape, y.shape)

# slice out a small number of points from the LHS of the domain
x_data = x[0:200:20]
y_data = y[0:200:20]
print(x_data.shape, y_data.shape)

plt.figure()
plt.plot(x, y, label="Exact solution")
plt.scatter(x_data, y_data, color="tab:orange", label="Training data")
plt.legend()
plt.show()
```

### Training a standard Neural Network

```python
# train standard neural network to fit training data
torch.manual_seed(123)
model = FCN(1,1,32,3)                                               # 1 input, 1 output, 32 hidden neurons, and 3 hidden layers
optimizer = torch.optim.Adam(model.parameters(),lr=1e-3)            # Adam optimizer with a learning rate of 10^(-3)
files = []
for i in range(1000):                                               #1000 iterations
    optimizer.zero_grad()                                           #Resets the gradients of all model parameters to zero. This is necessary because PyTorch accumulates gradients by default, and we want a fresh start for each iteration.
    yh = model(x_data)                                              # Model predicted output
    loss = torch.mean((yh-y_data)**2)# use mean squared error
    loss.backward()                                                 #Computes gradients for optimization.
    optimizer.step()                                                #Updates model parameters using gradients
    
    
    # plot the result as training progresses
    if (i+1) % 10 == 0: 
        
        yh = model(x).detach()
        
        plot_result(x,y,x_data,y_data,yh)
        
        file = "plots/nn_%.8i.png"%(i+1)
        plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
    
        if (i+1) % 500 == 0: plt.show()
        else: plt.close("all")
            
save_gif_PIL("nn.gif", files, fps=20, loop=0)
```

### Implement PINN

- add underlying differential equation to the loss function

Most of the cases we have prior understanding of what is going on {Physical Laws}
$$\min \frac{1}{N} \sum_{i}^{N} \left( u_{\text{NN}}(x_i; \theta) - u_{\text{true}}(x_i) \right)^2
+ \frac{1}{M} \sum_{j}^{M} \left( \left( m \frac{d^2}{dx^2} + \mu \frac{d}{dx} + k \right) u_{\text{NN}}(x_j; \theta) \right)^2$$

```python
x_physics = torch.linspace(0,1,30).view(-1,1).requires_grad_(True)# sample locations over the problem domain
mu, k = 2*d, w0**2

torch.manual_seed(123)                                                # Sets random seed for reproducibility
model = FCN(1,1,32,3)
optimizer = torch.optim.Adam(model.parameters(),lr=1e-4)
files = []
for i in range(20000):
    optimizer.zero_grad()
    
    # compute the "data loss"
    yh = model(x_data)
    loss1 = torch.mean((yh-y_data)**2)# use mean squared error
    
    # compute the "physics loss"
    yhp = model(x_physics)
    dx  = torch.autograd.grad(yhp, x_physics, torch.ones_like(yhp), create_graph=True)[0]# computes dy/dx
    dx2 = torch.autograd.grad(dx,  x_physics, torch.ones_like(dx),  create_graph=True)[0]# computes d^2y/dx^2
    physics = dx2 + mu*dx + k*yhp                                                        # computes the residual of the 1D harmonic oscillator differential equation
    loss2 = (1e-4)*torch.mean(physics**2)
    
    # backpropagate joint loss
    loss = loss1 + loss2                                             # add two loss terms together
    loss.backward()
    optimizer.step()
    
    
    # plot the result as training progresses
    if (i+1) % 150 == 0: 
        
        yh = model(x).detach()
        xp = x_physics.detach()
        
        plot_result(x,y,x_data,y_data,yh,xp)
        
        file = "plots/pinn_%.8i.png"%(i+1)
        plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
        
        if (i+1) % 3000 == 0: plt.show()
        else: plt.close("all")
            
save_gif_PIL("pinn.gif", files, fps=20, loop=0)
```
